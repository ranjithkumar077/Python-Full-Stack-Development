Scenario:
An online store has product descriptions. You need to highlight certain keywords or replace outdated terminology within these descriptions.

Problem/Task:
Given a product description string, you need to:
Count how many times a specific feature (e.g., "wireless") is mentioned.
Change all instances of an older term (e.g., "Bluetooth 4.0") to a new one ("Bluetooth 5.0").
Check if a crucial feature (e.g., "warranty") is present in the description.
Make a specific product name appear in all capital letters in a heading.

Example:
If a description is "This laptop has Bluetooth 4.0 and offers wireless connectivity.", you might want to replace "Bluetooth 4.0" and check for "warranty".

```python
Description = "This laptop has Bluetooth 4.0 and offers wireless connectivity"
#count
count = Description.count("wireless")
print("wireless_count :",count)
#Replace
Update = Description.replace("Bluetooth 4.0","bluetooth 5.0")
print("updated_description :",Update)
#check word
if "warranty" in Description:
    print("warranty is mentioned")
else:
    print("warranty is not mentioned")
    #upper case
product = "laptop"
capital = product.upper()
print("Uppercase :",capital)
```
