```python
#Upper and lower case
str6 = "Hello World"
print(str6.upper())  # Convert to uppercase
print(str6.lower())  # Convert to lowercase
```

```python
#Count
str7 = "hello world"
print(str7.count('o'))  # Count the number of occurrences of 'o'
```

```python
#Split
print(str7.split(" "))  # Split the string into a list of words
print(str7.split('o'))  # Split the string into a list of words, using 'o' as the delimiter
print(str7.split('l'))
```

```python
#Strip
str8 = "   Hello World   "
print(len(str8))
str9 = str8.rstrip()  # Remove leading and trailing whitespace
str9 = str8.lstrip() 
print(len(str9))  # Length after stripping whitespace
print(str9)
```

```python
#Replace
str10 = "Hello World"
str11 = str10.replace("World", "Python")  # Replace 'World' with 'Python'
print(str11)
#str10[0] = "h"  # This will cause an error because strings are immutable
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = list1  # This creates a reference to the same list
print(list1)
print(list2)
list1[1] = "cherry"
print(list1)
print(list2)
```

```python
list1 = [1, 'banana', 3, 4, 5]
list2 = list1.copy()  # This creates a shallow copy of the list
print(list1)
print(list2)
list1[1] = "cherry"
print(list1)
print(list2)
list2[1] = "grape"
print(list2)
```

```python
import copy as cp
list1 = ["a","b",["thor","Tony","natasha"],"c"]
list2 = cp.deepcopy(list1)  # This creates a deep copy of the list
print(list1)
print(list2)
list1[2][0] = "loki"
print(list1)
print(list2)
```

```python
task1 = input("Enter a task: ")
task2 = input("Enter another task: ")
task3 = input("Enter a third task: ")
tasks = [task1, task2, task3]
print(tasks)
tasks.sort()  # Sort the tasks alphabetically
print("Sorted_Tasks : ",tasks)
tasks.sort(reverse=True)  # Sort the tasks in reverse alphabetical order
print("Reverse_Sorted_Tasks : ",tasks)
tasks.insert(0," First_Read")
print(tasks)
print("After completing the task, remove it from the list")
tasks.pop(0)  # Remove the first task
tasks.pop(0)  # Remove the next task
print(tasks)




```
