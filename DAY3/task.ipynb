```python
Account_baalance = int(input("Enter your account balance"))
withdraw_limit = 15000
Minimum_balance = 1000
if(Account_baalance >= Minimum_balance):
    withdraw_amount = int(input("Enter the amount to withdraw"))
    if(withdraw_amount <= withdraw_limit and withdraw_amount <= Account_baalance - Minimum_balance):
            print("You are eligible to withdraw ")
            Account_baalance -= withdraw_amount
            print("Withdrawn amount:", withdraw_amount)
            print("Remaining balance:", Account_baalance)
    else:
            print("Insufficient balance")
else:
    print("Your account balance is below the minimum required balance.")
```
