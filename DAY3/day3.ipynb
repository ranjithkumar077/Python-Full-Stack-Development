Dictionary Methods


```python
d1 = {"key1": "value1", "key2": "value2"}
d2 = {"key3": "value3", "key4": "value4"}
#d3 = d1+d2 # it will give error because we cannot add two dictionaries using + operator
d3 = {**d1, **d2} # it will merge the two dictionaries d1 and d2 and create a new dictionary d3 ,
#packing and unpacking operator is used to merge two dictionaries and create a new dictionary
#d1.update(d2) # it will merge the two dictionaries d1 and d2 and update the dictionary d1 with the key-value pairs of d2
print(d3)
print(d1)
```

```python
d1_keys = d1.keys()#it will display th
print(d1_keys)
print(type(d1_keys))#it will display the type of keys which is dict_keys,so we need to convert it to a list to find the type of keys which is list
d1_keys_list = list(d1_keys)
print(type(d1_keys_list))#to find the type of the variable d1_keys_list which is a list and it is created by converting the dict_keys object d1_keys to a list using the list() function
```

```python
d1_values = tuple(d1.values())
print(d1_values)    
print(type(d1_values))
```

```python
print(d1["key1"]) # it will return the value of the key "key1" which is "value1"
print(d1["key3"]) # it will raise a KeyError because the key "key3" is not present in the dictionary d1
```

```python
print(d1.get("key1")) # it will return the value of the key "key1" which is "value1"
print(d1.get("key3")) # it will return None because the key "key3
print(d1.get("key3", "default value")) # it will return "default value" because the key "key3" is not present in the dictionary d1 and we have provided a default value to return when the key is not found in the dictionary
```

Escape character

```python
#str1 = 'this is thor's hammer' # it will give error because the string is not properly closed and it is also using single quotes which is conflicting with the single quotes used in the string
str1 = 'this is thor\'s hammer' # it will not give error because we have used escape character \ to escape the single quote in the string and it is also using single quotes to define the string
print(str1)
str2 = "this is thor's hammer" # it will not give error because we are using double quotes to define the string and it is also using single quotes in the string which is not conflicting with the double quotes used to define the string
print(str2)
```

```python
print("line1\nline2") # it will print line1 and line2 in two different lines because we have used the escape character \n to create a new line
print("line1\tline2") # it will print line1 and line2 with a tab space in between because we have used the escape character \t to create a tab space
print("line1\\line2") # it will print line1\line2 because we have used the escape character \\ to print a single backslash in the string
```
