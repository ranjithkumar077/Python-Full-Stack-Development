```python
str='this is thor\'s hammer'# \ is use for escape character 
print(str)
#horizontal tab
print('this is thor\t hammer')
#vertical tab
print('this is thor\v hammer')
#backspace
print('this is thor\b hammer')
#carriage return
print('this is thor\r hammer')
#form feed
print('this is thor\f hammer')
# \\ to print \
print('this is thor\\ hammer')
#\' to print '
print('this is thor\'s hammer')
#\" to print "
print('this is thor\"s hammer')
#\xhh to print character with hex value hh
print('this is thor\x20s hammer')# \x20 is space in hex
```
