```python
#how the interpreter works
#its starts with the source code and then it goes to the byte code and then it goes to the machine code and then it executes the code.

age = int(input('Enter your age: '))
if (age >= 18):
    print('You are eligible to vote')
else:
    print('You are not eligible to vote')
```

```python
#if-elif-else
#if-elif-else is used to check multiple conditions and execute the code accordingly.
marks = int(input('Enter your marks: '))
if (marks >= 90):
    print('You got A grade')
elif (marks >= 80):
    print('You got B grade')
elif (marks >= 60):
    print('You got C grade')
elif (marks >= 40):
    print('You got D grade')
else:
    print('You got F grade')
```

```python
#atm withdrawal .constriant balance and daily limit .update the balance 
balance = 50000
daily_limit = 15000
withdrawal_amount = int(input('Enter the amount to withdraw: '))
if (withdrawal_amount <= balance):
    if (withdrawal_amount <= daily_limit):
        balance -= withdrawal_amount
        print(f'Withdrawal successful,remaining balance is {balance}')
    else:
        print('withdrawal amount is more than daily limit')
print (balance)
```
