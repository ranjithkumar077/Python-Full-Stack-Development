Dictionary Methods 

```python
d1={"key1":"value1",
    "key2":"value2",
    }
d2={"key3":"value3",
    "key4":"value4",}
# d3=d1+d2 this not work because we can not use the + operator to add two dictionaries together
d4 = {**d1, **d2}# Dictionary unpacking operator ** to add items from one dictionary to another dictionary.
print(d4)
```

```python
#update method to add items to a dictionary
#extract keys 
d1_keys=list(d1.keys())# this will return a view object that displays a list of all the keys in the dictionary d1
print(d1_keys)
print(type(d1_keys))# this will return the type of the object d1_keys which is a list
#extract values
d1_values=list(d1.values())# this will return a view object that displays a list
print(d1_values)
print(type(d1_values))# this will return the type of the object d1_values which
#tuples convertion 
d2_keys_tuple=tuple(d2.keys())# this will return a view object that displays a list of all the keys in the dictionary d2 and convert it to a tuple
print(d2_keys_tuple)
print(type(d2_keys_tuple))# this will return the type of the object d2_keys_tuple which is a tuple
```
