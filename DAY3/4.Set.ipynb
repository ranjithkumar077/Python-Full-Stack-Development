```python
#set operations
s1={1,2,3}
s2={2,4,5}
s3=s1.union(s2)
print(s3)
s4=s1.intersection(s2)
print(s4)
s5=s1.difference(s2)#explains the items in s1 which are not in s2
print(s5)
```

```python
#set is unodered collection of unique items
#set is mutable
#we can add or remove items from set
#there is no concept of index in set
#we can not access items in set by index
list1=['a','b','c','d','e']
set=set(list1)
print(set)
#remove vs discard
set.remove('a')#if the item is not present in the set it will raise an error
print(set)
set.discard('b')#if the item is not present in the set it will not raise an error
print(set)
set.discard('z')#if the item is not present in the set it will not raise an error
print(set)
```
