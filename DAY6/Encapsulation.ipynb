# Encapsulation
Encapsulation in Python (and in object-oriented programming in general) means hiding the internal details of a class and only exposing what’s necessary through methods.

It protects the data (attributes) by restricting direct access.

Instead of accessing variables directly, you use getter and setter methods.

This ensures that the internal state of an object can’t be changed in unexpected ways.

```python
class class1():
    def __init__(self,a,b,c):
        self.a=a # public
        self._b=b# Protected
        self.__c=c # Private
obj=class1(10,20,30)
print(obj.a)
print(obj._b)
try:
    print(obj.__c)
except Exception:
    print("Cannot access in this way")
print(obj._class1__c)
        
```
