# Inheritance 

```python
# Single Level Inheritance 
#    Parents class(math1)====> Child class(Math2)
#           
class Math1():
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b

class Math2(Math1):
    def mul(self,a,b):
        return a*b

obj=Math2()
print(obj.add(2,5))
```

# MultiLevel Inheritance
This code demonstrates inheritance and method overriding in Python:

Math1 defines basic operations: addition and subtraction.

Math2 inherits from Math1 and adds multiplication.

Math3 inherits from Math2 but overrides the mul method to perform division instead.

When you create an object of Math3, it can use all methods from its parent classes (add, sub) and its own overridden mul.

```python
# Multi Level
class Math1():
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b

class Math2(Math1):
    def mul(self,a,b):
        return a*b
class Math3(Math2):
    def mul(self,a,b):
        return a/b


obj=Math3()
print(obj.add(2,5))
```

# Super Keyword
 it is used to access methods and properties of a parent class from within a child class

```python
# super() keyword is used to access methods and properties of a parent class from within a child class
class class1():
    def hello():
        print(" hello i am from claas 1")
    def sum(self,a,b):
        return a+b
    
class class2(class1):
    def Find_Mean(self,a,b):
        sum=super().sum(a,b)
        mean=sum/2
        return mean

obj=class2()
obj.Find_Mean(10,10)
```
