This task was about building a simple Bank class in Python to simulate basic banking operations. The class lets you:
=> Create a user account with a starting balance.
=> Check the current balance.
=> Deposit money into the account.
=> Withdraw money, with a safeguard against overdrawing.

```python

class Bank:
    def __init__(self, name, amount=5000):
        self.name = name
        self.amount = amount
        print(f"Name: {self.name}")
        print(f"Opening Balance: {self.amount}")

    def Check_Balance(self):
        print(f"Balance Amount: {self.amount}")

    def Add_Amount(self, add):
        self.amount += add
        print(f"Added amount: {add}")
        print(f"Current Balance: {self.amount}")

    def Withdrawal_Amount(self, withdraw):
        if withdraw > self.amount:
            print("Insufficient balance")
        else:
            self.amount -= withdraw
            print(f"Withdrawn amount: {withdraw}")
            print(f"Current Balance: {self.amount}")


user1 = Bank("Ranjith")
user1.Check_Balance()
user1.Add_Amount(5000)
user1.Withdrawal_Amount(200)
```
