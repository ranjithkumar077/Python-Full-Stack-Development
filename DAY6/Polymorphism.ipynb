# Polymorphism


```python
class SocialMedia():
    def share_pic(self,platform):
        print(f" Sharig in a{platform}")
    def post_stories(self,platform):
        print(f" posting a story in {platform}")
class insta(SocialMedia):
    pass
class fb(SocialMedia):
    pass
user1=insta()
user2=fb()
user1.share_pic(" insta")
user2.post_stories(" facebook")
```
