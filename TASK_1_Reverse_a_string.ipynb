```python
#2.	Check whether a string is a palindrome,
def is_palindrome(s):
    """Checks if the input string is a palindrome."""
    cleaned_s = ''.join(s.split()).lower()  # Remove spaces and convert to lowercase
    return cleaned_s == cleaned_s[::-1]
is_palindrome("A man a plan a canal Panama")  # True
is_palindrome("Hello")  # False 
#Check whether an integer is a palindrome.
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False
        original = x
        reversed_num = 0
        while x > 0:
            reversed_num = reversed_num * 10 + x % 10
            x //= 10
        return original == reversed_num
```

```python
#3.	Count the number of vowels and consonants in a string. 

def count_vowels_consonants(s):
    """Counts the number of vowels and consonants in the input string."""
    vowels = 'aeiouAEIOU'
    vowel_count = 0
    consonant_count = 0
    for char in s:
        if char.isalpha():  # Check if the character is an alphabet
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1
    return vowel_count, consonant_count
#leetode :345 Reverse Vowels of a String
```

```python
#6.	Count the frequency of each character in a string. 
def count_frequency(s):
    """Counts the frequency of each character in the input string."""
    frequency = {}
    for char in s:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency
print(count_frequency("hello world"))
print(count_frequency("hello worl"))
#leetcode : 387. First Unique Character in a String
```

```python
#16.	Check if two strings differ by exactly one character
def differ_by_one(s1, s2):
    if len(s1) != len(s2):
        return False
    mismatch_count = 0
    for c1, c2 in zip(s1, s2):
        if c1 != c2:
            mismatch_count += 1
            if mismatch_count > 1:
                return False
    return mismatch_count == 1

print(differ_by_one("cat", "bat"))   # True
print(differ_by_one("cat", "car"))   # True
print(differ_by_one("cat", "cats"))  # False
print(differ_by_one("cat", "dog"))   # False
```

```python
#18. Check if Two Strings are Rotations
def are_rotations(s1, s2):
    return len(s1) == len(s2) and s2 in (s1 + s1)

# Example
print(are_rotations("abcde", "deabc"))  # True
```

```python
# 20. Check if List is Monotonic
def is_monotonic(lst):
    return all(lst[i] <= lst[i+1] for i in range(len(lst)-1)) or \
           all(lst[i] >= lst[i+1] for i in range(len(lst)-1))

# Example
print(is_monotonic([1,2,2,3]))  # True
print(is_monotonic([5,4,4,2]))  # True
```

```python
# 22. Generate Permutations of a String
# python
from itertools import permutations

def string_permutations(s):
    return [''.join(p) for p in permutations(s)]

# Example
print(string_permutations("abc"))
```

```python
# 23. Longest Substring Without Repeating Characters
# python
def longest_unique_substring(s):
    seen = {}
    start = max_len = 0
    for i, c in enumerate(s):
        if c in seen and seen[c] >= start:
            start = seen[c] + 1
        seen[c] = i
        max_len = max(max_len, i - start + 1)
    return max_len

# Example
print(longest_unique_substring("abcabcbb"))  # 3
```

```python
#24. Longest Palindrome Substring
def longest_palindrome(s):
    def expand(l, r):
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1
            r += 1
        return s[l+1:r]
    res = ""
    for i in range(len(s)):
        res = max(res, expand(i, i), expand(i, i+1), key=len)
    return res

# Example
print(longest_palindrome("babad")) 
```

```python
# 25. Group Anagrams
from collections import defaultdict

def group_anagrams(words):
    groups = defaultdict(list)
    for w in words:
        groups[''.join(sorted(w))].append(w)
    return list(groups.values())

# Example
print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```
