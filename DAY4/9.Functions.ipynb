```python
#*args and **kwargs
#*args is used to pass a variable number of non-keyword arguments to a function. It allows you to handle more arguments than the number of formal parameters defined in the function.
def mul(*a  ):
    """Returns the multiplication of "a","b" and "c"."""
    m=1
    for i in a:
        m=m*i
    return m
print(mul(2,3,3,4,5,6))
```

```python
def fun1(a):
    return a*10
def fun2(fun,b):
    c=fun(b)
    return c
ans=fun2(fun1,6)
print(ans)
#explain the above code
# In the above code, we have two functions: `fun1` and `fun2`.
# 1. `fun1(a)`: This function takes a single argument `a` and returns the result of multiplying `a` by 10.
# 2. `fun2(fun, b)`: This function takes two arguments: `fun`, which is expected to be a function, and `b`, which is a value to be passed to that function. Inside `fun2`, we call the function `fun` with the argument `b` and store the result in variable `c`. Finally, `fun2` returns the value of `c`.
# When we call `fun2(fun1, 6)`, we are passing the function `fun1` as the first argument and the value `6` as the second argument. Inside `fun2`, it calls `fun1(6)`, which computes `6 * 10` and returns `60`. This value is then returned by `fun2` and stored in the variable `ans`, which is printed to the console, resulting in the output `60`.
```

```python
def say_hi():
    return "Hi"
def say_bye():
    return (say_hi(),"this is modified")
print(say_bye())
```

```python
def wrapper_function(fun):
    def modified_function():
        fun()
        print("This is modified")
    return modified_function
@wrapper_function
def say_hi():
    print("Hello")
# say_hi=wrapper_function(say_hi)#@wrapper_function
say_hi()
```

```python
# expected output:
# Hello
# 11
# done
def wrapper_function(fun):
    def sum_fun(a,b):
        print("hello")
        fun(a,b)
        print("done")
    return sum_fun
@wrapper_function
def add_fun(a,b):
    print(a+b)
add_fun(5,10)
```

Global & Local variables

```python
#Global and local variables
#Global variables are defined outside of any function and can be accessed and modified by any function in the program. Local variables, on the other hand, are defined within a function and can only be accessed and modified within that function.

b=10
def fun1():
    a=" inside fun1"
    global b
    print(a)
    print(b)

fun1()
print(b)
```
