```python
#Question 1: processing Finacial Transactions 
#    you are given a list of transactions for a day , some entries are invalid ( eg negative amount or non numeric value) you need to calculate the total valid transactions and count the number of invalid transactions
data = [150.50,-30.00, 200.00, -10.00, 50.05555, -100.00]
# task 1 : calculate total valid transactions and  sum the total valid transactions as final total 
#use for and if else statement to solve this problem
# for invalid we skip it 
total= 0
invalid_count = 0
for transaction in data:
    if transaction > 0  :
        total += transaction
    else:
        invalid_count += 1

print(f"Total valid transactions: {total}")
print(f"Number of invalid transactions: {invalid_count}")
```

BREAK AND CONTINUE  


```python
# Break statement in for loop.when the interrpret is the if condition is true it will break the loop and come out of the loop
list = ["a", "b", "c", "d","c","e"]
for i in list:
    if i == "e":
        break 
    print (i)
print("_" * 50)
#Continue statement in for loop when the interpret is the if condition is true it will skip the current iteration and move to the next iteration
for i in list:
    if i == "c":
        continue
    print(i)
```

```python
list = ["a", "b", "c", "d","c","e"]
for i in list:
    print (i)
    if i == "c":
        break 
    print (i)
```

```python
#  while loop vs for loop
# while loop is used when we want to repeat a block of code until a certain condition is met. It is more flexible than a for loop because it can be used in situations where the number of iterations is not known beforehand. A for loop, on the other hand, is typically used when we know the number of iterations or when we want to iterate over a sequence (like a list or a range). In summary, use a while loop when you need more control over the loop's execution and a for loop when you want to iterate over a sequence or a known number of times.
# while loop 
list =["thor", "hulk", "ironman"]
#print the list using while loop and length of the list
i=0
while i<len(list):
    print(list[i])
    i += 1 
```
