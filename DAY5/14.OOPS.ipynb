```python
class math():
    def __init__(self):# __init__ is the constructor which works automatically without calling or excution 
        print(" Class created ")
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b
obj=math()
obj1=math()# accessing the class math function as obj1 ,here object acts instance of the class .
ans1=obj1.add(2,8)# accessing or using  the add function.
print(ans1)
obj2=math()
ans2=obj2.sub(2,8)
print(ans2)

```

```python
class Person:
    def __init__(self, name, age):
        self.name = name   # store name
        self.age = age     # store age

    def greet(self):
        print(f"Hello, my name is {self.name} and age {self.age}")

# Create an object
p1 = Person("John", 36)

# Call the greet method
p1.greet()
```

```python
class dog():
    def bark(self):
        print("woff")
    def fetch(self,object):
        print(f"got {object}")
max=dog()
brownie=dog()
max.bark()
brownie.fetch("ball")
max.fetch("plate")
```

```python
class Dog():
    def __init__(self,name,breed):
        self.a=name
        self.b=breed
        print(f"{name}is Born!")
    def bark(self):
        print(f"name:{self.a}")
        print(f"breed :{self.b}")
my_obj1=Dog("max","germany")
my_obj1.bark()
print(my_obj1.a)
print(my_obj1.b)
my_obj2=Dog("tommy","golden")
```
